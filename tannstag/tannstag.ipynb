```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.draw as skid
from ipywidgets import *

%matplotlib inline
```

```python
z = 200

def draw(A):
    fb = np.zeros((z,z), dtype=np.uint8)

    N = 8
    R = 0.1
    r = 0.03

    C = 2 * np.pi * R
    X = (A / (2 * np.pi)) * C
    Y = 0
    
    for i in range(N):
        a = A + 2 * np.pi * (i / N)
        x = X + R * np.cos(a)
        y = Y + R * np.sin(a)

        xx, yy = skid.circle(z//2 + z*x, z//2 + z*y, z*r)
        fb[yy, xx] = 1

    plt.imshow(fb)
    
interact(draw, A=(-np.pi, np.pi, 0.001))
```

```python
z = 1000

fb = np.zeros((z,z), dtype=np.uint8)

def draw(A):
    N = 8
    R = 0.1
    r = 0.02

    C = 2 * np.pi * R
    X = (A / (2 * np.pi)) * C
    Y = 0
    
    for i in range(N):
        a = A + 2 * np.pi * (i / N)
        x = X + R * np.cos(a)
        y = Y + R * np.sin(a)

        xx, yy = skid.circle(z//2 + z*x, z//2 + z*y, z*r)
        fb[yy, xx] = 1

for a in np.linspace(-np.pi, np.pi, 400):
    draw(a)

plt.figure(figsize=(50,50))
plt.imshow(fb)
```
